### Diferença de desistências entre 2019 e 2020 por município
Importar os dados do enem de 2019 e 2020

```python
import pandas as pd
enem_2019 = pd.read_csv('MICRODADOS_ENEM_2019.csv', sep=';', encoding='latin-1')
```

```python
enem_2020 = pd.read_csv('MICRODADOS_ENEM_2020.csv', sep=';', encoding='latin-1')
```

Importar os códigos dos municípios

```python
codigos = pd.read_csv('RELATORIO_DTB_BRASIL_MUNICIPIO.csv')
codigos = codigos.loc[:,['Nome_UF','Nome_Município','Código Município Completo']]
codigos = codigos.rename(columns={'Nome_Município': 'municipio'})
codigos = codigos.rename(columns={'Código Município Completo': 'codigo'})
codigos = codigos.rename(columns={'Nome_UF': 'estado'})
codigos
```

#### Preparação dos dados de 2019

Selecionar somente participantes com o ano de conclusão igual ao ano de realização da prova

Para o cálculo de abstenção, conta-se pessoas que faltaram no primeiro dia de prova ou no segundo dia de prova

```python
import numpy as np
enem_EM_2019 = enem_2019[enem_2019['TP_ST_CONCLUSAO'] == 2] #Situação de conclusão do Ensino Médio
enem_abs_2019 = enem_EM_2019[np.logical_or(enem_EM_2019['TP_PRESENCA_CH'] == 0, enem_EM_2019['TP_PRESENCA_CN'] == 0)]
```

Agrupar as desistências de 2019 e dividir pela quantidade total de pessoas em 2019 pelo lugar que o participante estudou

```python
taxa_abs_2019 = enem_abs_2019.groupby(['CO_MUNICIPIO_PROVA']).size() / enem_EM_2019.groupby(['CO_MUNICIPIO_PROVA']).size() * 100
taxa_abs_2019 = taxa_abs_2019.to_frame().reset_index()
#Renomear as colunas como padronização
taxa_abs_2019 = taxa_abs_2019.rename(columns={'CO_MUNICIPIO_PROVA': 'codigo'})
taxa_abs_2019 = taxa_abs_2019.rename(columns={0: 'abstencao 2019'})
taxa_abs_2019 = taxa_abs_2019.sort_values(by='abstencao 2019',ascending=False)
taxa_abs_2019
```

```python
taxa_abs_2019 = taxa_abs_2019.dropna()
```

```python
taxa_abs_2019
```

#### Preparação dos dados de 2020

Selecionar somente participantes com o ano de conclusão igual ao ano de realização da prova

Para o cálculo de abstenção, conta-se pessoas que faltaram no primeiro dia de prova ou no segundo dia de prova

```python
enem_EM_2020 = enem_2020[enem_2020['TP_ST_CONCLUSAO'] == 2] #Situação de conclusão do Ensino Médio
enem_abs_2020 = enem_EM_2020[np.logical_or(enem_EM_2020['TP_PRESENCA_CH'] == 0, enem_EM_2020['TP_PRESENCA_CN'] == 0)]
```

```python
print("A quantidade de pessoas que fizeram o ENEM foi de",len(enem_EM_2019),", e a quantidade de pessoas que desistiram em 2019 foi",len(enem_abs_2019))
```

Agrupar as desistências de 2020 e dividir pela quantidade total de pessoas em 2020 pelo lugar que o participante estudou

```python
taxa_abs_2020 = enem_abs_2020.groupby(['CO_MUNICIPIO_PROVA']).size() / enem_EM_2020.groupby(['CO_MUNICIPIO_PROVA']).size() * 100
taxa_abs_2020 = taxa_abs_2020.to_frame()
taxa_abs_2020 = taxa_abs_2020.reset_index()
#Renomear as colunas como padronização
taxa_abs_2020 = taxa_abs_2020.rename(columns={'CO_MUNICIPIO_PROVA': 'codigo'})
taxa_abs_2020 = taxa_abs_2020.rename(columns={0: 'abstencao 2020'})
taxa_abs_2020
```

```python
taxa_abs_2020 = taxa_abs_2020.dropna()
```

```python
taxa_abs_2020 = taxa_abs_2020.sort_values(by='abstencao 2020',ascending=False)
taxa_abs_2020
```

#### Dataframe final
Junção dos dataframes de 2019 e 2020 pelo código do município

```python
df_taxas = pd.merge(taxa_abs_2020, taxa_abs_2019, how = 'inner', on = 'codigo')
df_taxas
```

Calcular a diferença entre os anos, ou seja, a taxa de abstenção de 2020 menos a taxa de abstenção de 2019

```python
df_taxas['diferenca'] = df_taxas['abstencao 2020'] - df_taxas['abstencao 2019']
df_taxas
```

```python
df_taxas = df_taxas.astype({"codigo": 'str'})
df_taxas.info()
```

```python
df_taxas
```

Padronização do código do município

```python
linha = 0
for i in df_taxas['codigo']:
    df_taxas.loc[linha,['codigo']]=i[0:7]
    linha += 1
df_taxas
```
