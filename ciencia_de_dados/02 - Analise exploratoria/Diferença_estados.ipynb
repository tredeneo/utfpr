### Diferença de desistências entre 2019 e 2020 por estado
Importar os dados do enem de 2019 e 2020

```python
import pandas as pd
df2019 = pd.read_csv('MICRODADOS_ENEM_2019.csv', sep=';', encoding='latin-1')
```

```python
df2020 = pd.read_csv('MICRODADOS_ENEM_2020.csv', sep=';', encoding='latin-1')
```

Selecionar somente participantes com o ano de conclusão igual ao ano de realização da prova

Para o cálculo de abstenção, conta-se pessoas que faltaram no primeiro dia de prova ou no segundo dia de prova

```python
import numpy as np
df_enem2019 = df2019[df2019['TP_ST_CONCLUSAO'] == 2] #Situação de conclusão do Ensino Médio
df_abstencao2019 = df_enem2019[np.logical_or(df_enem2019['TP_PRESENCA_CH'] == 0, df_enem2019['TP_PRESENCA_CN'] == 0)]
```

```python
df_enem2020 = df2020[df2020['TP_ST_CONCLUSAO'] == 2] #Situação de conclusão do Ensino Médio
df_abstencao2020 = df_enem2020[np.logical_or(df_enem2020['TP_PRESENCA_CH'] == 0, df_enem2020['TP_PRESENCA_CN'] == 0)]
```

#### Preparação dos dados de 2019

```python
df_abstencao2019.info()
```

```python
df_enem2019.info()
```

Agrupar as desistências de 2019 e dividir pela quantidade total de pessoas em 2019 pelo lugar que o participante estudou

```python
abstencao_estados2019 = df_abstencao2019.groupby(['SG_UF_PROVA']).size() / df_enem2019.groupby(['SG_UF_PROVA']).size() * 100
abstencao_estados2019 = abstencao_estados2019.to_frame().reset_index()
#Renomear as colunas como padronização
abstencao_estados2019 = abstencao_estados2019.rename(columns={'SG_UF_PROVA': 'estado'})
abstencao_estados2019 = abstencao_estados2019.rename(columns={0: 'abstencao 2019'})
abstencao_estados2019.info()
```

Ordenar por taxa de abstenção

```python
abstencao_estados2019 = abstencao_estados2019.sort_values(by='abstencao 2019',ascending=False)
abstencao_estados2019
```

#### Preparação dos dados de 2020

Agrupar as desistências de 2020 e dividir pela quantidade total de pessoas em 2020 pelo lugar que o participante estudou

```python
abstencao_estados2020 = df_abstencao2020.groupby(['SG_UF_PROVA']).size() / df_enem2020.groupby(['SG_UF_PROVA']).size() * 100
abstencao_estados2020 = abstencao_estados2020.to_frame().reset_index()
#Renomear as colunas como padronização
abstencao_estados2020 = abstencao_estados2020.rename(columns={'SG_UF_PROVA': 'estado'})
abstencao_estados2020 = abstencao_estados2020.rename(columns={0: 'abstencao 2020'})
abstencao_estados2020
```

Ordenar por taxa de abstenção

```python
abstencao_estados2020 = abstencao_estados2020.sort_values(by='abstencao 2020',ascending=False)
abstencao_estados2020
```

#### Dataframe final
Junção dos dataframes de 2019 e 2020 por estado

```python
df_taxas = pd.merge(abstencao_estados2020, abstencao_estados2019, how = 'inner', on = 'estado')
```

Alteração de UF para nome dos estados 

```python
UF_dict = { 'AC':'Acre',
            'AL':'Alagoas',
            'AP':'Amapá',
            'AM':'Amazonas',
            'BA':'Bahia',
            'CE':'Ceará',
            'DF':'Distrito Federal',
            'ES':'Espírito Santo',
            'GO':'Goiás',
            'MA':'Maranhão',
            'MT':'Mato Grosso',
            'MS':'Mato Grosso do Sul',
            'MG':'Minas Gerais',
            'PA':'Pará',
            'PB':'Paraíba',
            'PR':'Paraná',
            'PE':'Pernambuco',
            'PI':'Piauí',
            'RJ':'Rio de Janeiro',
            'RN':'Rio Grande do Norte',
            'RS':'Rio Grande do Sul',
            'RO':'Rondônia',
            'RR':'Roraima',
            'SC':'Santa Catarina',
            'SP':'São Paulo',
            'SE':'Sergipe',
            'TO':'Tocantins'}
```

```python
df_taxas["estado"] = df_taxas["estado"].replace(UF_dict)
# Estado como idex
df_taxas.set_index('estado',inplace = True)
df_taxas
```

Calcular a diferença entre os anos, ou seja, a taxa de abstenção de 2020 menos a taxa de abstenção de 2019

```python
df_taxas['diferenca'] = df_taxas['abstencao 2020'] - df_taxas['abstencao 2019']
df_taxas
```
