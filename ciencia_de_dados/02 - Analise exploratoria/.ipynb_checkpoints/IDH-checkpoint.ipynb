### Análise por município

Importar os dados do IDHM 

```python
import pandas as pd
import seaborn as sn
idh = pd.read_csv('IDH_2010.csv')
idh
```

Processar os dados

```python
idh = idh.rename(columns={'Município': 'municipio'})
idh = idh.rename(columns={'Código do Município': 'codigo'})
idh = idh.loc[:,['IDHM','IDHM Educação','IDHM Renda','IDHM Longevidade','municipio','codigo']]
idh["municipio"] = idh["municipio"].str.title()
idh
```

Exportar o dataframe

```python
#idh.to_csv("idh.csv")
```

Junção com os dados do ENEM

```python
enem_municipios = pd.read_csv('taxas_abstencao_municipio.csv')
enem_municipios = enem_municipios.drop(columns=['Unnamed: 0'])
enem_municipios
```

```python
#Padronizar os códigos dos municípios
enem_municipios = enem_municipios.astype({"codigo": 'str'})
linha = 0
for i in enem_municipios['codigo']:
    enem_municipios.loc[linha,['codigo']]=i[:-1]
    linha += 1
enem_municipios
```

```python
idh = idh.astype({"codigo": 'str'})
idh
```

```python
enem_idh = pd.merge(idh, enem_municipios, how = 'inner', on = 'codigo')
enem_idh
```

Correlação entre as variáveis

```python
correlation = enem_idh.corr()
plot = sn.heatmap(correlation, annot = True, fmt=".3f", linewidths=.6)
```

### Análise por estado
Importar dados do IDH

```python
idh_estados = pd.read_csv('idhm_estados.csv')
idh_estados
```

Junção com dados do ENEM

```python
enem_estados = pd.read_csv('taxas_abstencao_estado.csv')
enem_estados
```

```python
enem_idh_estado = pd.merge(idh_estados, enem_estados, how = 'inner', on = 'estado')
enem_idh_estado
```

Correlação entre as variáveis

```python
correlation = enem_idh_estado.corr()
plot = sn.heatmap(correlation, annot = True, fmt=".3f", linewidths=.6)
```
