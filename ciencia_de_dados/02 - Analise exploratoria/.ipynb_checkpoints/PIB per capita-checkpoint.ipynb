### Análise por município
Importar dados do PIB

```python
import pandas as pd
import seaborn as sn
pib = pd.read_csv('PIB-dos-Municípios-base-de-dados-2010-2019.csv')
pib
```

Processar os dados 

```python
# Renomear colunas
pib = pib.rename(columns={'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)': 'PIB'})
pib = pib.rename(columns={'Nome do Município': 'municipio'})
pib = pib.rename(columns={'Código do Município': 'codigo'})
# PIB mais recente
pib = pib[pib['Ano']==2019]
# Dados necessários
pib = pib.loc[:,['municipio','PIB','codigo']]
pib.info()
```

```python
#PIB em float
pib['PIB'] = pib['PIB'].str.replace(',',' ')
pib['PIB'] = pib['PIB'].str.replace(" ","")
pib['PIB'] = pib['PIB'].str.replace('(',"")
pib['PIB'] = pib['PIB'].str.replace(')',"")
pib['PIB'] = pib['PIB'].astype(float)
pib['PIB']
```

```python
pib
```

Exportar dataframe do PIB

```python
#pib.to_csv("pib.csv")
```

Realizar a junção com os dados do ENEM pelo código do município

```python
enem_municipios = pd.read_csv('taxas_abstencao_municipio.csv')
enem_municipios = enem_municipios.drop(columns=['Unnamed: 0'])
enem_municipios
```

```python
enem_pib = pd.merge(pib, enem_municipios, how = 'inner', on = 'codigo')
enem_pib
```

Correlação entre as variáveis

```python
correlation = enem_pib.corr()
plot = sn.heatmap(correlation, annot = True, fmt=".3f", linewidths=.6)
```

### Análise por estado
Importar dados do PIB por estado

```python
pib_estado = pd.read_csv('PIB_per_capita_final.csv')
pib_estado
```

Junção com os dados do ENEM

```python
enem_estados = pd.read_csv('taxas_abstencao_estado.csv')
enem_estados
```

```python
enem_pib_estado = pd.merge(pib_estado, enem_estados, how = 'inner', on = 'estado')
enem_pib_estado
```

Correlação entre as variáveis

```python
correlation = enem_pib_estado.corr()
plot = sn.heatmap(correlation, annot = True, fmt=".1f", linewidths=.6).set_title('Correlação entre as variáveis PIB per Capita e o número de abstenção por estado')
```
