### Regressão linar por município

```python
import pandas as pd
import seaborn as sn
from statsmodels.formula.api import ols
```

#### PIB per capita
Importar PIB per capita e processar os dados

```python
pib = pd.read_csv('pib.csv')
pib = pib.drop(columns=['Unnamed: 0'])
pib = pib.sort_values(by=['municipio'], ascending=True)
pib
```

#### IDHM
Importar IDHM per capita e processar os dados

```python
idh = pd.read_csv('idh.csv') 
idh = idh.drop(columns=['Unnamed: 0'])
idh = idh.sort_values(by=['municipio'], ascending=True)
idh
```

#### Dados sociais
Importar os dados sociais, para análise será utilizada apenas a densidade demográfica

```python
socio = pd.read_csv('socio_stats_wide_cities.csv')
socio
```

```python
socio = socio.rename(columns={'Name': 'municipio'})
socio = socio.rename(columns={'Densidade Demográfica (hab/km²)': 'densidade'})
socio = socio.rename(columns={'CodMun': 'codigo'})
socio = socio.loc[:,['municipio','densidade','codigo']]
socio = socio.sort_values(by=['municipio'], ascending=True)
```

```python
socio
```

#### Taxas de abstenção dos municípios
Importar as taxas de abstenção dos municípios e processar os dados

```python
taxas = pd.read_csv('taxas_abstencao_municipio.csv')
taxas.set_index('Unnamed: 0',inplace = True)
taxas
```

#### Dados de COVID-19
Importar os dados de COVID-19 e processar os dados

```python
covid = pd.read_csv('df_projeto_final_covid_e_escola.csv', sep=';')
covid = covid.loc[:,['Total_Alunos','Relacao_cadastrados','Casos_per_capita','Mortes_per_capita','Relacao_casos_mes','Relacao_obitos_mes','Diferenca_aprovacao','Diferenca_reprovacao','Diferenca_desistencia','codigo']]
covid
```

```python
covid = covid.astype({"codigo": 'str'})
linha = 0
for i in covid['codigo']:
    covid.loc[linha,['codigo']]=i[:-3]
    linha += 1
covid
```

#### Junção dos dados socioeconômicos com os dados de COVID-19
Juntar os dados pelo código do município

```python
df1 = pd.merge(pib, taxas, how = 'inner', on = 'codigo')
df1
```

```python
df1 = df1.astype({"codigo": 'str'})
linha = 0
for i in df1['codigo']:
    df1.loc[linha,['codigo']]=i[:-1]
    linha += 1
df1
```

```python
df2 = pd.merge(socio, idh, how = 'inner', on = 'codigo')
df2 = df2.astype({"codigo": 'str'})
df2
```

```python
df = pd.merge(df1, df2, how = 'inner', on = 'codigo')
df
```

```python
df = df.loc[:,['PIB','IDHM','densidade','diferenca','IDHM Renda','codigo']]
df
```

Importar os códigos dos municíos para realizar a junção e obter os estados

```python
codigos = pd.read_csv('codigos.csv')
codigos
```

```python
codigos = codigos.astype({"codigo": 'str'})
linha = 0
for i in codigos['codigo']:
    codigos.loc[linha,['codigo']]=i[:-1]
    linha += 1
codigos
```

```python
df = pd.merge(df, codigos, how = 'inner', on = 'codigo')
df
```

```python
df = pd.merge(df, covid, how = 'inner', on = 'codigo')
df
```

```python
df.set_index('municipio',inplace = True)
df = df.loc[:,['PIB','IDHM','densidade','diferenca','estado','Total_Alunos','Relacao_cadastrados','Casos_per_capita','Mortes_per_capita','Relacao_casos_mes','Relacao_obitos_mes','Diferenca_aprovacao','Diferenca_reprovacao','Diferenca_desistencia']]
```

```python
df
```

Selecionar apenas os municíos com cinco ou mais estudantes

```python
df_mun_maiores = df[(df['Total_Alunos'] >= 5)]
df_mun_maiores
```

#### Modelo final
Agrupar por estado com São Paulo como referência, utilizando PIB, IDHM, densidade, Diferenca_aprovacao e Relacao_casos_mes como variáveis 

```python
model1 = ols("diferenca ~ C(estado, Treatment(reference='São Paulo')) + IDHM + PIB + Casos_per_capita + Diferenca_reprovacao", data=df_mun_maiores)
response1 = model1.fit()
response1.summary()
```

```python
model2 = ols("diferenca ~ C(estado, Treatment(reference='São Paulo')) + IDHM + PIB + Mortes_per_capita + Diferenca_reprovacao", data=df_mun_maiores)
response2 = model2.fit()
response2.summary()
```
