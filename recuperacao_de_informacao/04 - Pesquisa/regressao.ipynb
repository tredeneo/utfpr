```python
import pandas
import statsmodels.formula.api as modelos #ols , logit
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(style='ticks')
def mostrar_grafico(nome_x,nome_y,df,titulo):
    #sns.scatterplot(data=dataframe, x=nome_x, y=nome_y)
    #sns.lineplot(data=dataframe, x=nome_x, y=modelo.fittedvalues, color='red')
    sns.regplot(data=df, x=nome_x, y=nome_y, ci=None,
            line_kws={'color': 'red'}, scatter_kws={'color': 'blue'})
    plt.title(titulo)
    plt.xlabel(nome_x)
    plt.ylabel(nome_y)
    plt.show()
```

```python
from dataclasses import dataclass,fields
@dataclass
class Grupos:
    mutiplayer=[
        'ARK: Survival Evolved',
        "Garry's Mod",
        'Dead by Daylight',
        'Sid Meier’s Civilization® VI',
        'Rocket League®',
        'Human: Fall Flat',
        'Insurgency: Sandstorm',
        'Rust',
        'Grand Theft Auto V',
        "PLAYERUNKNOWN'S BATTLEGROUNDS",
    ]
    coop = [
        'ASTRONEER',
        "Don't Starve Together",
        'Divinity: Original Sin 2 - Definitive Edition',
        'MONSTER HUNTER: WORLD',
        'Stardew Valley',
        'Terraria',
        'Left 4 Dead 2',
    ]
    singleplayer = [
        'Factorio',
        'Euro Truck Simulator 2',
        'Slay the Spire',
        'Subnautica',
        'The Elder Scrolls V: Skyrim Special Edition',
        'RimWorld',
        'RESIDENT EVIL 2',
    ]
    fps = [
        'Grand Theft Auto V',
        "PLAYERUNKNOWN'S BATTLEGROUNDS",
        'Left 4 Dead 2',
        'RESIDENT EVIL 2',
    ]
    rpg = [
        'Divinity: Original Sin 2 - Definitive Edition',
        'MONSTER HUNTER: WORLD',
        'Terraria',
        'Slay the Spire',
        'The Elder Scrolls V: Skyrim Special Edition',
    ]
    survival = [
        'ARK: Survival Evolved',
        'Rust',
        'ASTRONEER',
        "Don't Starve Together",
        'Stardew Valley',
        'RimWorld',
        'Subnautica',
    ]
    competitive = [
        'Rocket League®',
        'Dead by Daylight',
        'Insurgency: Sandstorm',
        "PLAYERUNKNOWN'S BATTLEGROUNDS",
    ]
```

```python
colunas = ["hour_played",
           "recommendation","title",
          "polaridade_bayes"]
original = pandas.read_parquet("../Dados/tratados.parquet",columns=colunas)
```

```python
dados = original[:]
dados = pandas.get_dummies(dados,columns=["recommendation"],dtype=int)
dados = dados.drop(['recommendation_Not Recommended'], axis=1)
dados = dados.rename(columns={'recommendation_Recommended': 'recommend','polaridade_bayes': 'polaridade'})

dados.dtypes
```

# hipótese 1.a
* Os reviews que recomendam o jogo possuem um sentimento mais positivo
    em comparação com os que não recomendam

```python
tmp = dados[["hour_played","polaridade","recommend"]]
sns.heatmap(tmp.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

# hipótese 1.b:
* Jogadores que investem mais horas no jogo tendem a ter um sentimento
mais positivo sobre o jogo, e vice-versa

```python
model = modelos.ols("hour_played ~ polaridade", data=dados)
response = model.fit()
response.summary()
```

```python
mostrar_grafico("hour_played","polaridade",dados,"regressão analises individuais")
```

```python
model = modelos.ols("hour_played ~ polaridade + C(recommend)", data=dados)
response = model.fit()
response.summary()
```

# hipótese 1.C
* review com sentimento positivo e com muitas horas jogadas está recomendando o jogo

```python
model = modelos.logit("recommend ~ hour_played + polaridade", data=dados)
response = model.fit()
response.summary()
```

# hipótese 2.a
* A qualidade percebida do jogo (medida pela porcentagem de recomendação
dos reviews) pode influenciar a motivação dos jogadores para continuar
jogando e, portanto, uma maior média de quantidade de horas jogadas

```python
agregacoes = {"polaridade":"mean","hour_played":"median",
             "recommend":"sum"}
#tmp = 
agrupado = dados.groupby('title').agg(agregacoes).reset_index()
agrupado['Contagem'] = dados.groupby('title').size().reset_index(name='Contagem')['Contagem']

agrupado['porcent_recomend'] = (agrupado['recommend'] * 100) / agrupado['Contagem']
agrupado
```

```python
tmp = agrupado[["hour_played","polaridade","porcent_recomend"]]
sns.heatmap(tmp.corr(), annot=True, cmap="coolwarm_r")
plt.title('Matriz de Correlação agrupado')
plt.show()
```

```python
model = modelos.ols("hour_played ~ porcent_recomend + polaridade", data=agrupado)
response = model.fit()
response.summary()
```

```python
mostrar_grafico("hour_played","porcent_recomend",agrupado,"regressão jogos agrupados")
mostrar_grafico("hour_played","polaridade",agrupado,"regressão jogos agrupados")
```
