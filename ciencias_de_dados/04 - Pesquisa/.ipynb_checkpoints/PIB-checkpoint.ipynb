```python
import pandas as pd
from statsmodels.formula.api import ols
dados_estados = pd.read_csv('dados_estados_pib.csv')
dados_estados.set_index('estado',inplace = True)
dados_estados
```

```python
dados_estados.plot.scatter('PIB', 'abstencao')
```

```python
model = ols("PIB ~ abstencao", data=dados_estados)
response = model.fit()
response.summary()
```

```python
dados_municipios = pd.read_csv('dados_municipios_pib.csv')
dados_municipios.set_index('Nome_Município',inplace = True)
dados_municipios
```

```python
dados_municipios.plot.scatter('PIB per capita', 'abstencao')
```

```python
dados_municipios = dados_municipios.rename(columns={"PIB per capita":"PIB"})
```

```python
model2 = ols("PIB ~ abstencao", data=dados_municipios)
response2 = model2.fit()
response2.summary()
```
