```python
import pandas as pd
from statsmodels.formula.api import ols
dados_municipios_idh = pd.read_csv('dados_municipios_idh.csv')
dados_municipios_idh.set_index('Nome_Município',inplace = True)
dados_municipios_idh
```

```python
dados_municipios_idh.plot.scatter('IDHM', 'abstencao')
```

```python
model = ols("IDHM ~ abstencao", data=dados_municipios_idh)
response = model.fit()
response.summary()
```

```python
dados_estados_idh = pd.read_csv('dados_estados_idh.csv')
dados_estados_idh.set_index('estado',inplace = True)
dados_estados_idh
```

```python
dados_estados_idh.plot.scatter('IDHM', 'abstencao')
```

```python
model = ols("IDHM ~ abstencao", data=dados_estados_idh)
response = model.fit()
response.summary()
```
